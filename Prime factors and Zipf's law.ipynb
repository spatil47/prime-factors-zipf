# Prime factors and Zipf's law

```julia
using Primes
```

```julia
using Plots
```

```julia
prime_factors = [factor(x) for x in 1:4000] 
```

```julia
prime_factors_collected = merge(+, prime_factors...)
```

```julia
scatter(
    prime_factors_collected;
    title = "Frequency distribution of prime factors",
    ylabel = "Frequency",
    xlabel = "Prime factor",
    legend = :none,
    yaxis = :log,
    xaxis = :log
)
```
